```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# 데이터 불러오기
subway_congestion_data = pd.read_csv("서울교통공사_지하철혼잡도정보_20231231.csv", encoding='cp949')

# 혼잡도 데이터 선택
congestion_data = subway_congestion_data.iloc[:, 6:]

# NaN 값 처리
congestion_data_scaled = congestion_data.fillna(congestion_data.mean())

# Inf 값 처리
congestion_data_scaled.replace([np.inf, -np.inf], np.nan, inplace=True)
max_vals = congestion_data_scaled.max().max()
congestion_data_scaled.fillna(max_vals, inplace=True)

# 데이터 표준화
scaler = StandardScaler()
congestion_data_scaled = scaler.fit_transform(congestion_data_scaled)

# K-means 모델 생성 및 학습
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(congestion_data_scaled)

# 클러스터링 결과를 데이터프레임에 추가
subway_congestion_data['Cluster'] = kmeans.labels_

# 클러스터별 혼잡도 시각화
plt.figure(figsize=(12, 6))
for cluster_label in range(3):
    cluster_data = subway_congestion_data[subway_congestion_data['Cluster'] == cluster_label].iloc[:, 6:-1].mean()
    plt.plot(cluster_data, label=f'Cluster {cluster_label + 1}')

plt.title('클러스터별 평균 혼잡도 패턴')
plt.xlabel('시간')
plt.ylabel('평균 혼잡도')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 데이터 불러오기
subway_congestion_data = pd.read_csv("서울교통공사_지하철혼잡도정보_20231231.csv", encoding='cp949')

# 혼잡도 데이터 선택
congestion_data = subway_congestion_data.iloc[:, 6:]

# Z 점수 계산
z_scores = (congestion_data - congestion_data.mean()) / congestion_data.std()

# 이상치 임계값 설정 (예를 들어, Z 점수가 3보다 크면 이상치로 간주)
threshold = 3

# 이상치 탐지
outliers = (z_scores > threshold).any(axis=1)

# 이상치 출력
print("이상치 개수:", outliers.sum())
print("이상치 비율:", outliers.mean())
print("이상치 인덱스:", outliers[outliers].index.tolist())
```
